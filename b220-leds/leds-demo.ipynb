```python
from pynq import Overlay
```

```python
ol = Overlay('./overlay/leds.bit')
```

```python
#ol.ip_dict
```

```python
leds_ip = ol.ip_dict['axi_gpio_0']
```

```python
from pynq.lib import AxiGPIO
from time import sleep
```

```python
leds =  AxiGPIO(leds_ip).channel1
leds.setdirection('out')
leds.setlength(4)
```

```python
for i in range(16):
    leds.write(i, 0xf)
    sleep(0.2)
```
