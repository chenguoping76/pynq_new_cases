# 演示使用opencv2获取USB接口的web camera图像  


## 请使用B220 pynq 2.7 镜像完成验证

## 确认b220 正确识别 usb webcamera  


```python
# 第一步引入cv2 库，并使用cv2.VideoCapture(0) 获得视频输入设备
import cv2
videoIn = cv2.VideoCapture(0)
```

```python
ret, frame_vga = videoIn.read()
```

```python
# 使用matplotlib 库的pyplt显示原始图片
import matplotlib.pyplot as plt
# 注意cv2 采集图像为BGR格式，需要使用cvtColor转换为RGB，以确保颜色正确
plt.imshow(cv2.cvtColor(frame_vga,cv2.COLOR_BGR2RGB))
```
