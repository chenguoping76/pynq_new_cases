```python
from pynq.overlays.base import BaseOverlay
base = BaseOverlay("base.bit")
```

```python
%%microblaze base.ARDUINO
#include "gpio.h"
#include "timer.h"
void blink_d2(void){
    gpio arduino_pin_d2 = gpio_open(2);
    gpio_set_direction(arduino_pin_d2, GPIO_OUT);
    for(int i = 0; i< 10; i++) {
        gpio_write(arduino_pin_d2, 1);
        delay_ms(500);
        gpio_write(arduino_pin_d2, 0);
        delay_ms(500);
    }
}
    
```
