```python
from pynq.overlays.logictools import LogicToolsOverlay
```

```python
logictools = LogicToolsOverlay('./overlay/logictools.bit')
```

```python
boolean_generator = logictools.boolean_generator
```

```python
function = ['LD2 = PB3 ^ PB0']
```

```python
boolean_generator.setup(function)
```

```python
boolean_generator.run()
```
